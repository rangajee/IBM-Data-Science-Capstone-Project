```python
import folium
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Unique sites (name + coords)
sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Center the map on the mean coordinate
center = [sites['Lat'].mean(), sites['Long'].mean()]
m = folium.Map(location=center, zoom_start=5, tiles="OpenStreetMap")

# Helpful plugins
MarkerCluster(name="Launch Sites").add_to(m)
MousePosition(position='topright', prefix="Lat/Lon:").add_to(m)

# Add a pin, a small circle, and a label for each site
for _, r in sites.iterrows():
    name, lat, lon = r['Launch Site'], float(r['Lat']), float(r['Long'])

    folium.Marker(
        [lat, lon],
        popup=folium.Popup(name, max_width=250),
        tooltip=name,
        icon=folium.Icon(color="blue", icon="rocket", prefix="fa")
    ).add_to(m)

    folium.Circle([lat, lon], radius=750, color="blue", fill=True, fill_opacity=0.1).add_to(m)

    folium.Marker(
        [lat, lon],
        icon=DivIcon(icon_size=(150, 36), icon_anchor=(0, 0),
                     html=f'<div style="font-size:12px;font-weight:600;white-space:nowrap;">{name}</div>')
    ).add_to(m)

folium.LayerControl().add_to(m)
m  # display the map in the notebook
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math
from folium import Marker, PolyLine

# Haversine formula to calculate distance between two (lat, lon) points in km
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # radius of Earth in km
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c

# Example: distances for multiple sites
sites = [
    {'name': 'CCAFS LC-40', 'lat': 28.562302, 'lon': -80.577356, 'coast_lat': 28.563, 'coast_lon': -80.567},
    {'name': 'VAFB SLC-4E', 'lat': 34.632834, 'lon': -120.610746, 'coast_lat': 34.633, 'coast_lon': -120.600},
]

for site in sites:
    distance = calculate_distance(site['lat'], site['lon'], site['coast_lat'], site['coast_lon'])
    folium.PolyLine(
        [[site['lat'], site['lon']], [site['coast_lat'], site['coast_lon']]],
        color='blue', weight=2, opacity=0.8
    ).add_to(site_map)
    folium.Marker(
        [site['coast_lat'], site['coast_lon']],
        icon=DivIcon(
            icon_size=(150, 36), icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{site["name"]}<br>{distance:.2f} km</b></div>'
        ),
    ).add_to(site_map)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import math, folium
from folium.features import DivIcon

# Haversine (km)
def haversine_km(lat1, lon1, lat2, lon2):
    R=6371.0
    from math import radians, sin, cos, asin, sqrt
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    return 2*R*asin(sqrt(sin((lat2-lat1)/2)**2 + cos(lat1)*cos(lat2)*sin((lon2-lon1)/2)**2))

# Pick the site you’re working on
SITE = "CCAFS LC-40"   # or "VAFB SLC-4E", "KSC LC-39A", etc.

site_lat = float(spacex_df.loc[spacex_df['Launch Site']==SITE, 'Lat'].iloc[0])
site_lon = float(spacex_df.loc[spacex_df['Launch Site']==SITE, 'Long'].iloc[0])

# >>> Replace these with the coordinates you read using MousePosition <<<
features = [
    ("Coastline", 28.563000, -80.567000),
    ("Railway",   28.572640, -80.567770),
    ("Highway",   28.562080, -80.577950),
    ("City: Cape Canaveral", 28.392200, -80.607700),
]

for label, tlat, tlon in features:
    d = haversine_km(site_lat, site_lon, tlat, tlon)

    # line from site to target
    folium.PolyLine([[site_lat, site_lon], [tlat, tlon]],
                    color='blue', weight=2, opacity=0.8).add_to(site_map)

    # marker with distance label at the target
    folium.Marker(
        [tlat, tlon],
        icon=DivIcon(
            icon_size=(180, 40), icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;color:#d35400;"><b>{label}<br>{d:.2f} km</b></div>'
        )
    ).add_to(site_map)

site_map  # re-display the updated map
```

**Are launch sites in close proximity to railways?**  
**Yes.**  
All SpaceX launch sites (such as CCAFS LC-40, KSC LC-39A, and VAFB SLC-4E) are located within a few kilometers of railways.  
Railways are essential for transporting large rocket components, engines, and fuel to the sites efficiently and safely.  
Typical distances are **less than 5 km**.

---

**Are launch sites in close proximity to highways?**  
**Yes.**  
Each site is located near major highways or access roads—again, usually within a few kilometers.  
Highways provide road transport for personnel, smaller parts, and logistical support.  
These roads are designed to handle **heavy, oversized loads** related to rocket transport.

---

**Are launch sites in close proximity to the coastline?**  
**Absolutely.**  
All launch sites are intentionally built very close to the coastline, often **within 2 km**.  
This allows rockets to launch eastward over the ocean, ensuring that debris or rocket stages fall safely into the water rather than over land.  
The coast also provides large, open safety zones with minimal population exposure.

---

**Do launch sites keep certain distance away from cities?**  
**Yes, by design.**  
All launch sites are located well away from large urban areas—typically **10–30 km** from the nearest city.  
This separation reduces risk in case of launch failures, minimizes noise/vibration impacts, and ensures safety for residents.  
For example:  
- *Cape Canaveral* sites are about **20 km from the city of Cape Canaveral**.  
- *Vandenberg SLC-4E* is about **15 km from Lompoc, California**.
